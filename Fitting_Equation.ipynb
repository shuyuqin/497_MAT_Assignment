```
def V(x, y, a, b, h, k, T, alpha, gamma):
    """
    Return the Voigt line shape at x with Lorentzian component HWHM gamma
    and Gaussian component HWHM alpha.

    """
    
    sigma = alpha / np.sqrt(2 * np.log(2))

    x_p = (x-h)*np.cos(T)-(y-k)*np.sin(T)
    y_p = (x-h)*np.sin(T)-(y-k)*np.cos(T)

    u = (x_p/a)**2 + (y_p/b)**2

    return np.real(wofz((u + 1j*gamma)/sigma/np.sqrt(2))) / sigma\
                                                           /np.sqrt(2*np.pi)
```

```
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xx, yy = np.meshgrid(x, y, sparse=True)
```

```
data = V(xx,yy,4,1,1, 3, .5, .3,.1)
```
